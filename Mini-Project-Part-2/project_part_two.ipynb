```python
# Dependencies
import pandas as pd
```

```python
# load CSV
new_coders = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
new_coders_df = pd.read_csv(new_coders, low_memory ='False')
new_coders_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_codersel_df = new_coders_df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_codersel_df
```

```python
# Change reading 0 and 1 to No and Yes, respectively
codersel_df = new_codersel_df.replace({0:'No', 1:'Yes'})
codersel_df.head(50)
```

```python
# Extract rows for only those who attended a bootcamp
attended_df = codersel_df.loc[codersel_df['AttendedBootcamp']=='Yes']
attended_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
BootcampsAttendees = pd.DataFrame(new_coders_df["BootcampName"].value_counts())
BootcampsAttendees.reset_index(inplace=True)
BootcampsAttendees.columns = ["BootcampName", "Number of Attendees"]
BootcampsAttendees_df = BootcampsAttendees.loc[(BootcampsAttendees["Number of Attendees"] >= 5)]
BootcampsAttendees_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommenders_df = attended_df.replace({"No": 0, "Yes": 1})

recommenders_df = pd.DataFrame(new_coders_df.groupby("BootcampName")["BootcampRecommend"].sum())
recommenders_df.reset_index(inplace=True)
recommenders_df.columns = ["BootcampName", "Number of Recommenders"]

recommenders_df.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_data = pd.merge(BootcampsAttendees, recommenders_df, on="BootcampName")
merge_data.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merge_data["Percentage of Recommended"] = round(merge_data["Number of Recommenders"]/merge_data["Number of Attendees"], 3) *100
merge_data.head()
```

```python
# Export to excel and remove index
merge_data.to_csv("output/magagi_output.csv", index=False, header=True)
```
